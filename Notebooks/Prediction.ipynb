```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier


X = data_mice
y = data_mice["hospital_death"]

rf = RandomForestClassifier(n_estimators=100, max_depth=3,
                                 bootstrap=True, n_jobs=-1,
                                 random_state=0)
rf.fit(X, y)

feature_imp = pd.Series(rf.feature_importances_, 
                        index=data_mice.columns).sort_values(ascending=False)

print('Feature importances: ', rf.feature_importances_)
print(sns.barplot(x=feature_imp, y=feature_imp.index))
plt.xlabel('Feature Importance Score', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.title("Visualizing Important Features", fontsize=15, pad=15)
```

```python
from sklearn.feature_selection import SelectFromModel

selector = SelectFromModel(rf, threshold=0.05)
features_important = selector.fit_transform(X, y)

print('Data with initial features')
print(pd.DataFrame(X, columns=data_mice.columns).head())

print('Data with selected features')
print(pd.DataFrame(features_important).head())
```

```python
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
# define dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=10, random_state=7)
# define the pipeline
steps = [('pca', PCA(n_components=10)), ('m', LogisticRegression())]
model = Pipeline(steps=steps)
# evaluate model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
target_count = data_mice.hospital_death.value_counts()
print('Class 0:', target_count[0])
print('Class 1:', target_count[1])
print('Proportion:', round(target_count[0] / target_count[1], 2), ': 1')


target_count.plot(kind='bar', title='Count (target)');
```

```python
# Class count
count_class_0, count_class_1 = data_mice.hospital_death.value_counts()

# Divide by class
df_class_0 = data_mice[data_mice['hospital_death'] == 0]
df_class_1 = data_mice[data_mice['hospital_death'] == 1]
```

```python
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
df_test_over = pd.concat([df_class_0, df_class_1_over], axis=0)

print('Random over-sampling:')
print(df_test_over.hospital_death.value_counts())

df_test_over.hospital_death.value_counts().plot(kind='bar', title='Count (target)');
```

```python
from sklearn.ensemble import RandomForestRegressor
df1=df_test_over.drop(['hospital_death'], axis=1)
model = RandomForestRegressor(random_state=1, max_depth=10)
df=pd.get_dummies(df1)
model.fit(df,df_test_over.hospital_death)
```

```python
import numpy as np
```

```python
features = df.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-10:]  # top 10 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
